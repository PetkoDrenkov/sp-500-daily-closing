```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sp = pd.read_csv("/home/petko/Desktop/SPY_close_price_5Y.csv")
sp["log_returns"] = np.log(sp["Close"]) - np.log(sp["Close"].shift(1))
sp.info()
```

```python
# plotting time series
plt.plot(sp["Date"],sp["Close"])
plt.xticks(np.linspace(0,1255,10), rotation=50)
plt.title("Price")
plt.show()
```

```python
# plotting logarithmic returns: log(price_on_given_day) - log(price_of_previous_day)
plt.plot(sp["Date"],sp["log_returns"])
plt.xticks(np.linspace(0,1255,10), rotation=50)
plt.title("Logarithmic Returns")
plt.show()
```
